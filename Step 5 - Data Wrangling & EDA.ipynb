## **EDA**

**Loan Paid vs Not Paid Count:** The bar chart below shows that about 80% of the loans in the dataset were paid back and about 20% were not paid back.

```python
counts = df["loan_paid_back"].value_counts().sort_index()
total = counts.sum()

# Convert index to strings for cleaner x-axis labels
x_labels = counts.index.astype(str)

# Choose colors (0 = red, 1 = green)
colors = ['red', 'green']

plt.figure(figsize=(6, 4))

bars = plt.bar(x_labels, counts.values, color=colors)

# Add percentage labels above bars
for i, bar in enumerate(bars):
    height = bar.get_height()
    pct = (height / total) * 100
    plt.text(
        bar.get_x() + bar.get_width() / 2,
        height,
        f"{pct:.1f}%",
        ha="center",
        va="bottom",
        fontsize=10
    )

plt.title("Loan Paid vs Not Paid")
plt.xlabel("loan_paid_back (0 = Not Paid, 1 = Paid)")
plt.ylabel("Count")
plt.tight_layout()
plt.show()
```

**Descriptive Statistics Comparison:** Below are two tables of descriptive statistics, one for loans that were paid back and one for loans that weren't. In general, I didn't see a lot of significant differences between the two. The one thing I did notice is that the mean credit score for the loans that were pack back was notably higher than the loans that weren't paid back, 686.2 vs 651.4.

```python
paid_back_df = df[df["loan_paid_back"] == 1]
not_paid_back_df = df[df["loan_paid_back"] == 0]

# Summary statistics for loans that were paid back
paid_back_summary = round(paid_back_df.describe(),1)

# Summary statistics for loans that were not paid back
not_paid_back_summary = round(not_paid_back_df.describe(),1)

paid_back_summary
```

**Heatmap Comparison:** I generated the Heatmap below to identify numerical features that are strongly correlated and identify features that had any correlation with the target (loan_paid_back). The three feature pairs with the strongest correlations were monthly_income/annual_income, installment/loan_amount and num_of_delinquencies/delinquency_history. The four features that had the strongest correlation with the target on an absolute basis were debt_to_income_ratio, credit_score, interest_rate and delinquency_history.

```python
numerical_features = df.select_dtypes(include=['int64', 'float64'])
plt.figure(figsize=(12, 8))
sns.heatmap(numerical_features.corr(), annot=True, cmap='coolwarm',fmt='.2f')
plt.title("Correlation Heatmap")
plt.show()
```

**Histograms of Correlated Features:** The below histograms were generated for the four features with the strongest correlation with the target. The interest_rate showed a normal distribution, debt_to_income_ratio and delinquency_history showed a right-skewed distribution and credit_score showed a slightly left-skewed distribution.

```python
numeric_features = [
    "debt_to_income_ratio",
    "credit_score",
    "interest_rate",
    "delinquency_history"
]

plt.figure(figsize=(8, 6))

for i, col in enumerate(numeric_features, 1):
    plt.subplot(2, 2, i)
    plt.hist(df[col].dropna(), bins=30)  # default color
    plt.title(f"Distribution of {col}")
    plt.xlabel(col)
    plt.ylabel("Frequency")

plt.tight_layout()
plt.show()
```

**Pie Chart for Loan Purpoose**
This Pie Chart shows a breakout of the distribution based on loan purpose. There are eight different purposes for taking out a loan based on this dataset. At about 40%, Debt Consolidation is clearly the most common reason. Vacation is the leat common at just 3%.

```python
counts = df["loan_purpose"].value_counts()

plt.figure(figsize=(7, 7))

# Create the pie chart
plt.pie(
    counts.values,
    labels=counts.index,
    autopct=lambda p: f"{p:.1f}%\n({int(p * counts.sum() / 100)})",
    startangle=140,
    wedgeprops={"edgecolor": "white"}
)

plt.title("Loan Purpose Distribution")
plt.tight_layout()
plt.show()
```

**Average Loan Paid-Back Rate:** For all of the categorical features, I created bar charts to show the average target value for each category. For four of the features, there appeared to be little to no difference among the categories. For two of the features (empoyment_status, grade_subgrade) there was noticeable differences among the categories. For empoyment_status, the Unemployed and Student categories showed a significantly lower Average Payback Rate than the other categories. For grade_subgrade, the rate showed a gradual increase as the loan grade became less risky.

```python
# Identify categorical columns
cat_cols = df.select_dtypes(include=['object']).columns.tolist()

for col in cat_cols:
    plt.figure(figsize=(6, 4))

    # Group by category and compute mean target value
    avg_paid_back = df.groupby(col)['loan_paid_back'].mean().sort_values()

    # Create bar plot
    avg_paid_back.plot(kind='bar')

    plt.title(f"Average Loan Paid-Back Rate by {col}")
    plt.ylabel("Average loan_paid_back")
    plt.xlabel(col)
    plt.xticks(rotation=45)
    plt.tight_layout()

    plt.show()
```

**Bar Charts for Categorical Features:** I generated the two bar charts below for employment_status and grade_subgrade because those were the only two categorical features that showed any noticeable differences among categories for the Average Loan Paid-Back Rate analysis above. The employment_satus bar chart shows that Employed is by far the most common. The grade_subgrade chart shows that the C and D grades are the most common.

```python

categorical_features = ["employment_status", "grade_subgrade"]

for col in categorical_features:
    plt.figure(figsize=(6, 4))
    df[col].value_counts().sort_index(ascending=False).plot(kind="bar")

    plt.title(f"Distribution of {col}")
    plt.xlabel(col)
    plt.ylabel("Count")
    plt.xticks(rotation=45, ha="right")
    plt.tight_layout()
    plt.show()
```
